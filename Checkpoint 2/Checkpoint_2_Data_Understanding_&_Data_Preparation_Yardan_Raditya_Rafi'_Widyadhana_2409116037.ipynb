3. Outliers

```python
results = []

cols = df.select_dtypes(include=['float64', 'int64'])

for col in cols:
  q1 = df[col].quantile(0.25)
  q3 = df[col].quantile(0.75)
  iqr = q3 - q1
  lower_bound = q1 - 1.5*iqr
  upper_bound = q3 + 1.5*iqr
  outliers = df[(df[col] < lower_bound) | (df[col] > upper_bound)]
  percent_outliers = (len(outliers)/len(df))*100
  results.append({'Kolom': col, 'Persentase Outliers': percent_outliers})

# Dataframe dari list hasil
results_df = pd.DataFrame(results)
results_df.set_index('Kolom', inplace=True)
results_df = results_df.rename_axis(None, axis=0).rename_axis('Kolom', axis=1)

# Tampilkan dataframe
display(results_df)
```

Hasil dari kode diatas menunjukkan bahwa terdapat nilai yang jauh berbeda pada bagian ParentalEducation, Music, dan Volunteering. Oleh karena itu diperlukan penanganan lebih lanjut terhadap outliers.

Dikarenakan presentase yang tidak terlalu tinggi maka dapat melakukan imputasi

```python
columns_to_impute = ["ParentalEducation", "Music", "Volunteering"]

for col in columns_to_impute:
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    # Menggunakan .loc[] agar tidak muncul SettingWithCopyWarning
    df.loc[:, col] = df[col].clip(lower=lower_bound, upper=upper_bound).astype(df[col].dtype)
```

Melakukan pengecekan ulang

```python
results = []

cols = df.select_dtypes(include=['float64', 'int64'])

for col in cols:
  q1 = df[col].quantile(0.25)
  q3 = df[col].quantile(0.75)
  iqr = q3 - q1
  lower_bound = q1 - 1.5*iqr
  upper_bound = q3 + 1.5*iqr
  outliers = df[(df[col] < lower_bound) | (df[col] > upper_bound)]
  percent_outliers = (len(outliers)/len(df))*100
  results.append({'Kolom': col, 'Persentase Outliers': percent_outliers})

# Dataframe dari list hasil
results_df = pd.DataFrame(results)
results_df.set_index('Kolom', inplace=True)
results_df = results_df.rename_axis(None, axis=0).rename_axis('Kolom', axis=1)

# Tampilkan dataframe
display(results_df)
```

Hasil dari kode di atas menunjukkan bahwa setelah dilakukan penanganan outliers, tidak ada lagi nilai yang jauh berbeda dalam dataset. Ini berarti proses penyesuaian telah berhasil dilakukan, sehingga data menjadi lebih bersih dan siap untuk analisis lebih lanjut.

5. Construct Data

berdasarkan data yang ada dan studi kasus yang sedang dianalisis. Kolom baru dibuat berdasarkan StudyTimeWeekly, Tutoring, ParentalSupport, dan Extracurricular untuk menentukan strategi pembelajaran siswa.

**Kriteria Learning Strategy**
1. Intensive → StudyTimeWeekly tinggi (≥10 jam) + Tutoring atau ParentalSupport tinggi.
2. Balanced → StudyTimeWeekly sedang (5-10 jam) + ada Extracurricular.
3. Minimalist → StudyTimeWeekly rendah (<5 jam) + minim dukungan (Tutoring & ParentalSupport rendah).

```python
def classify_learning_strategy(row):
    if row["StudyTimeWeekly"] >= 10 and (row["Tutoring"] == 1 or row["ParentalSupport"] >= 3):
        return "Intensive"
    elif 5 <= row["StudyTimeWeekly"] < 10 and row["Extracurricular"] == 1:
        return "Balanced"
    else:
        return "Minimalist"

# Menambahkan kolom baru "LearningStrategy"
df["LearningStrategy"] = df.apply(classify_learning_strategy, axis=1)

# Menampilkan beberapa data sebagai contoh
print(df[["StudentID", "StudyTimeWeekly", "Tutoring", "ParentalSupport", "Extracurricular", "LearningStrategy"]].head())
```
